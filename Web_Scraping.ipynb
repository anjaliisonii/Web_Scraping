```python
import requests 
from bs4 import BeautifulSoup
import webbrowser 
URL = "https://www.insiderscreener.com/en/explore/au" 
# webbrowser.open(URL)
html = requests.get(URL) 
soup = BeautifulSoup(html.text, 'html.parser')


```

```python
# soup.find('table', class_ ="table table-sm text-dark")
```

```python
# soup.find_all('th')
```

```python
table = soup.select_one("div.table-responsive-md table")
```

```python
rows_title= [th.text.strip() for th in table.select("thead th")]
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns= rows_title)
df
```

```python
column_data=table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data=row.find_all('td')
    ind=[data.text.strip() for data in row_data]
    length=len(df)
    df.loc[length]=ind
```

```python
df
```

```python
df.to_json('file1.json', orient = 'split', compression = 'infer')
```
